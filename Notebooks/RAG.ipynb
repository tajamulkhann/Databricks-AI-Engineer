Now that we have the `PyPDF2` library installed, we can write a function to search for the name in a specified PDF file. For demonstration, I will search in `/content/C.F.P. School.pdf`. You can change the `pdf_file_path` to any of the other available PDF files if needed.

````python
import PyPDF2

def search_name_in_pdf(pdf_file_path, name_to_search):
    found = False
    try:
        with open(pdf_file_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text = page.extract_text()
                if text and name_to_search.lower() in text.lower():
                    print(f"'{name_to_search}' found on page {page_num + 1} of {pdf_file_path}")
                    found = True
    except Exception as e:
        print(f"Error processing PDF {pdf_file_path}: {e}")

    if not found:
        print(f"'{name_to_search}' not found in {pdf_file_path}")


# Specify the PDF file path and the name to search
pdf_file_path = '/content/Taratala Ad. Vidyalaya (Pr.Sch) Rm-2.pdf'
name_to_search = 'Yasmin'

# Call the function to search for the name
search_name_in_pdf(pdf_file_path, name_to_search)
````

This code defines a function `search_name_in_pdf` that takes the PDF file path and the name to search as input. It opens the PDF, iterates through each page, extracts the text, and checks if the target name (case-insensitive) is present. It then prints whether the name was found and on which page.

````python
def getLLMResponse(prompt,model="databricks-claude-3-7-sonnet"):
  completion = client.chat.completions.create(
    model=model,
    messages= [
      {        "role": "system",
        "content": "You are a helpful assistant."
      },
      {
        "role": "user",
        "content":prompt
      } ],
  )
  return completion.choices[0].message.content

````
